```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
random_var = np.random.laplace(0, 1, 100)
plt.hist(random_var)
plt.axvline(random_var.mean(), color='black', linestyle='solid', linewidth=2)
plt.axvline(random_var.mean()+random_var.std(), color='black', linestyle='dashed', linewidth=2)
plt.axvline(random_var.mean()-random_var.std(), color='black', linestyle='dashed', linewidth=2)
plt.show()
```

```python
random_var = np.random.chisquare(4, 100)
plt.hist(random_var)
plt.axvline(random_var.mean(), color='black', linestyle='solid', linewidth=2)
plt.axvline(random_var.mean()+random_var.std(), color='black', linestyle='dashed', linewidth=2)
plt.axvline(random_var.mean()-random_var.std(), color='black', linestyle='dashed', linewidth=2)
plt.show()
```
